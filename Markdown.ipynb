Introduction
-------------------------------------------------------------------------------------

Le notebook Jupyter est une application web, donc intimement liée au format [HTML].
Selon [Wikipedia][HTML]:

[HTML]: https://en.wikipedia.org/wiki/HTML


> **Hypertext Markup Language** (**HTML**) is the standard [markup language] 
> for creating [web pages] and [web applications].
> With [Cascading Style Sheets] (CSS) and [JavaScript] 
> it forms a triad of cornerstone technologies for the [World Wide Web].  

[markup language]: https://en.wikipedia.org/wiki/Markup_language
[web pages]: https://en.wikipedia.org/wiki/Web_page
[web applications]: https://en.wikipedia.org/wiki/Web_application
[Cascading Style Sheets]: https://en.wikipedia.org/wiki/Cascading_Style_Sheets
[Javascript]: https://en.wikipedia.org/wiki/JavaScript
[World Wide Web]: https://en.wikipedia.org/wiki/World_Wide_Web


Par exemple, le fragment de HTML qui correspond au texte ci-dessus sur Wikipedia:

    <p><b>Hypertext Markup Language</b> (<b>HTML</b>) is the standard 
    <a href="/wiki/Markup_language" title="Markup language">markup language</a> 
    for creating <a href="/wiki/Web_page" title="Web page">web pages</a> 
    and <a href="/wiki/Web_application" title="Web application">web applications</a>. 
    With <a href="/wiki/Cascading_Style_Sheets" title="Cascading Style Sheets">
    Cascading Style Sheets</a> (CSS) and 
    <a href="/wiki/JavaScript" title="JavaScript">JavaScript</a> 
    it forms a triad of cornerstone technologies for the 
    <a href="/wiki/World_Wide_Web" title="World Wide Web">World Wide Web</a>.</p>


En raison des balises (ou *markup*) HTML, ce fragment est difficile à déchiffrer.  
**Vous ne voudriez probablement pas écrire "à la main" un tel document n'est-ce pas ?**

Une Alternative
---------------------------------------------------------------------------------

[Markdown] est un format texte, plus lisible, qui s'inspire des conventions de l'e-mail.  
A titre d'exemple, le fragment ci-dessus, décrit cette fois en Markdown:

    > **Hypertext Markup Language** (**HTML**) is the standard [markup language] 
    > for creating [web pages] and [web applications].  
    > With [Cascading Style Sheets] (CSS) and [JavaScript] 
    > it forms a triad of cornerstone technologies for the [World Wide Web].  

    [markup language]: https://en.wikipedia.org/wiki/Markup_language
    [web pages]: https://en.wikipedia.org/wiki/Web_page
    [web applications]: https://en.wikipedia.org/wiki/Web_application
    [Cascading Style Sheets]: https://en.wikipedia.org/wiki/Cascading_Style_Sheets
    [Javascript]: https://en.wikipedia.org/wiki/JavaScript
    [World Wide Web]: https://en.wikipedia.org/wiki/World_Wide_Web

Le notebook Jupyter permet d'inclure des cellules Markdown et prend en charge la conversion vers HTML.

[Markdown]: https://en.wikipedia.org/wiki/Markdown

Découvrir Markdown
-------------------------------------------------------------------------------------------------

### Un point de départ

Le site de [CommonMark], une spécification standard de Markdown.  
En particulier un [rappel de la syntaxe][commonmark-help] et 
leur leçon pour [apprendre Markdown en 10 minutes][commonmark-tutorial].

### Usages de Markdown

Markdown est aujourd'hui utilisé dans de nombreux contextes en dehors de Jupyter, par exemple:

 - [sites de questions et réponses][Stack Overflow], 
 
 - [plate-formes de publication web][Ghost],
 
 - [services web d'hébergement de logiciels][GitHub].
